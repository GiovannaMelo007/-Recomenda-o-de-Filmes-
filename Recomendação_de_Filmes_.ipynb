```python
dados_filmes = pd.read_csv('/content/movies.csv',usecols=['movieId','title'],dtype={'movieId': 'int32', 'title': 'str'})
```

```python
dados_filmes#dados_filmes Ã© o DataFrame que contÃ©m informaÃ§Ãµes sobre os filmes
```

```python
avaliacao_df=pd.read_csv('/content/ratings.csv',usecols=['userId', 'movieId', 'rating'],#especificando quais colunas serÃ£o lidas e incluÃ­das no DataFrame
    dtype={'userId': 'int32', 'movieId': 'int32', 'rating': 'float32'})
```

```python
avaliacao_df.head()# avaliacao_df Ã© o DataFrame que contÃ©m informaÃ§Ãµes sobre as avaliaÃ§Ãµes de filmes
```

juntando as bases

```python
df = pd.merge(avaliacao_df,dados_filmes,on='movieId')# A junÃ§Ã£o Ã© realizada com base na coluna 'movieId'
df
```

```python
df_recomendacao=df.pivot_table(index="userId", columns="title", values='rating').fillna(0)#substituir nan por 0
# index="userId": Define a coluna "userId" como o Ã­ndice da tabela pivÃ´
# columns="title": Define os tÃ­tulos dos filmes como as colunas da tabela pivÃ´
# values='rating': Usa os valores da coluna 'rating' para preencher a tabela pivÃ´
df_recomendacao#fica uma matriz esparsa
```

https://medium.com/brasil-ai/knn-k-nearest-neighbors-1-e140c82e9c4e


vizinhos proximos Ã© um modelo= dada uma determinada distancia  de outros  objetos  ele calcula uma similaridade de um vizinho


```python
from sklearn.neighbors import NearestNeighbors# Importando a classe NearestNeighbors do mÃ³dulo neighbors da biblioteca scikit-learn
```

 Cria uma instÃ¢ncia do modelo k-NN com as configuraÃ§Ãµes especificadas


 metric = 'cosine': Define a mÃ©trica de distÃ¢ncia como cosseno. Isso significa que o modelo calcularÃ¡ a similaridade entre os pontos usando a distÃ¢ncia cosseno.



```python
modelo_knn = NearestNeighbors(metric = 'cosine')
modelo_knn.fit(df_recomendacao)
```

```python
df_recomendacao.iloc[0]
```

 Acessa a primeira linha do DataFrame,  retorna uma sÃ©rie que contÃ©m os valores de todas as colunas para o primeiro usuÃ¡rio (primeira linha). Cada valor na sÃ©rie representa a avaliaÃ§Ã£o (ou rating) do respectivo filme para esse usuÃ¡rio

```python
df_recomendacao.shape#nÃºmero de linhas e colunas
```

```python
query_index = 230#Ã­ndice do usuÃ¡rio que serÃ¡ usado como consulta
distancia, indices_vizinhos = modelo_knn.kneighbors(df_recomendacao.iloc[query_index,:].values.reshape(1, -1), n_neighbors = 3)
```

query_index: Ã‰ o Ã­ndice do usuÃ¡rio que serÃ¡ usado como consulta para encontrar os vizinhos mais prÃ³ximos.

df_recomendacao.iloc[query_index,:].values.reshape(1, -1): Seleciona as avaliaÃ§Ãµes do usuÃ¡rio especificado (query_index) e remodela esses valores para ter o formato necessÃ¡rio para a consulta do modelo.

reshape(1, -1) Ã© usado para garantir que os dados sejam fornecidos ao modelo como uma matriz 2D, onde a primeira dimensÃ£o Ã© 1 (indicando uma Ãºnica amostra) e a segunda dimensÃ£o Ã© automaticamente calculada com base no nÃºmero de caracterÃ­sticas (neste caso, o nÃºmero de filmes).

n_neighbors = 3: Especifica que queremos encontrar os 3 vizinhos mais prÃ³ximos do usuÃ¡rio especificado.

```python
usuario =df_recomendacao.index[query_index]

for i in range(0, len(distancia.flatten())):
    if i == 0:
        print('Recomendando para usuÃ¡rio {0}:\n'.format(usuario))
    else:
        print('UsuÃ¡rio recomendado {0}: {1}, distÃ¢ncia de {2}:'.format(i, df_recomendacao.index[indices_vizinhos.flatten()[i]], distancia.flatten()[i]))
```

```python
idx_recomendado =df_recomendacao.index[indices_vizinhos.flatten()[1]]
idx_recomendado
```

```python
df_recomendacao.loc[230]
```

```python
recomendacao = df_recomendacao.loc[idx_recomendado].to_frame()
base = df_recomendacao.loc[usuario].to_frame()
```

df_recomendacao.loc[idx_recomendado]: Seleciona a linha com o Ã­ndice idx_recomendado do DataFrame df_recomendacao, que contÃ©m as avaliaÃ§Ãµes dos filmes recomendados para esse usuÃ¡rio.
df_recomendacao.loc[usuario]: Seleciona a linha com o Ã­ndice usuario do DataFrame df_recomendacao, que contÃ©m as avaliaÃ§Ãµes dos filmes na base de dados desse usuÃ¡rio.

```python
novos_titulos = pd.merge(base,recomendacao,on='title',how='inner').sort_values(by=idx_recomendado, ascending=False)
novos_titulos = novos_titulos[(novos_titulos[idx_recomendado] > 0) & (novos_titulos[usuario] == 0)].reset_index()
novos_titulos
```
