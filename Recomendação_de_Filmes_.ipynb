```python
dados_filmes = pd.read_csv('/content/movies.csv',usecols=['movieId','title'],dtype={'movieId': 'int32', 'title': 'str'})
```

```python
dados_filmes#dados_filmes é o DataFrame que contém informações sobre os filmes
```

```python
avaliacao_df=pd.read_csv('/content/ratings.csv',usecols=['userId', 'movieId', 'rating'],#especificando quais colunas serão lidas e incluídas no DataFrame
    dtype={'userId': 'int32', 'movieId': 'int32', 'rating': 'float32'})
```

```python
avaliacao_df.head()# avaliacao_df é o DataFrame que contém informações sobre as avaliações de filmes
```

juntando as bases

```python
df = pd.merge(avaliacao_df,dados_filmes,on='movieId')# A junção é realizada com base na coluna 'movieId'
df
```

```python
df_recomendacao=df.pivot_table(index="userId", columns="title", values='rating').fillna(0)#substituir nan por 0
# index="userId": Define a coluna "userId" como o índice da tabela pivô
# columns="title": Define os títulos dos filmes como as colunas da tabela pivô
# values='rating': Usa os valores da coluna 'rating' para preencher a tabela pivô
df_recomendacao#fica uma matriz esparsa
```

https://medium.com/brasil-ai/knn-k-nearest-neighbors-1-e140c82e9c4e


vizinhos proximos é um modelo= dada uma determinada distancia  de outros  objetos  ele calcula uma similaridade de um vizinho


```python
from sklearn.neighbors import NearestNeighbors# Importando a classe NearestNeighbors do módulo neighbors da biblioteca scikit-learn
```

 Cria uma instância do modelo k-NN com as configurações especificadas


 metric = 'cosine': Define a métrica de distância como cosseno. Isso significa que o modelo calculará a similaridade entre os pontos usando a distância cosseno.



```python
modelo_knn = NearestNeighbors(metric = 'cosine')
modelo_knn.fit(df_recomendacao)
```

```python
df_recomendacao.iloc[0]
```

 Acessa a primeira linha do DataFrame,  retorna uma série que contém os valores de todas as colunas para o primeiro usuário (primeira linha). Cada valor na série representa a avaliação (ou rating) do respectivo filme para esse usuário

```python
df_recomendacao.shape#número de linhas e colunas
```

```python
query_index = 230#índice do usuário que será usado como consulta
distancia, indices_vizinhos = modelo_knn.kneighbors(df_recomendacao.iloc[query_index,:].values.reshape(1, -1), n_neighbors = 3)
```

query_index: É o índice do usuário que será usado como consulta para encontrar os vizinhos mais próximos.

df_recomendacao.iloc[query_index,:].values.reshape(1, -1): Seleciona as avaliações do usuário especificado (query_index) e remodela esses valores para ter o formato necessário para a consulta do modelo.

reshape(1, -1) é usado para garantir que os dados sejam fornecidos ao modelo como uma matriz 2D, onde a primeira dimensão é 1 (indicando uma única amostra) e a segunda dimensão é automaticamente calculada com base no número de características (neste caso, o número de filmes).

n_neighbors = 3: Especifica que queremos encontrar os 3 vizinhos mais próximos do usuário especificado.

```python
indices_vizinhos
```

```python
len(distancia.flatten())#flatten(): Achatando a matriz de distâncias em uma única dimensão
```

```python
usuario =df_recomendacao.index[query_index]

for i in range(0, len(distancia.flatten())):
    if i == 0:
        print('Recomendando para usuário {0}:\n'.format(usuario))
    else:
        print('Usuário recomendado {0}: {1}, distância de {2}:'.format(i, df_recomendacao.index[indices_vizinhos.flatten()[i]], distancia.flatten()[i]))
```

```python
idx_recomendado =df_recomendacao.index[indices_vizinhos.flatten()[1]]
idx_recomendado
```

```python
df_recomendacao.loc[230]
```

```python
recomendacao = df_recomendacao.loc[idx_recomendado].to_frame()
base = df_recomendacao.loc[usuario].to_frame()
```

df_recomendacao.loc[idx_recomendado]: Seleciona a linha com o índice idx_recomendado do DataFrame df_recomendacao, que contém as avaliações dos filmes recomendados para esse usuário.
df_recomendacao.loc[usuario]: Seleciona a linha com o índice usuario do DataFrame df_recomendacao, que contém as avaliações dos filmes na base de dados desse usuário.

```python
novos_titulos = pd.merge(base,recomendacao,on='title',how='inner').sort_values(by=idx_recomendado, ascending=False)
novos_titulos = novos_titulos[(novos_titulos[idx_recomendado] > 0) & (novos_titulos[usuario] == 0)].reset_index()
novos_titulos
```
